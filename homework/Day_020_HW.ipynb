```python
nrow = 1000
ncol = 3
upper = 1
lower = -1
"""
Your Code Here
"""
matrix = matrix = (upper - lower) * np.random.random((nrow, ncol)) + lower

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seaborn 進階 Heatmap
grid = sns.PairGrid(
    data=plot_data, 
    size=3, 
    diag_sharey=False, 
    hue='index', 
    vars=[x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha=0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)


plt.show()
```
