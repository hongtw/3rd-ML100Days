- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
import numpy as np

np.random.seed(9487)
%matplotlib inline

iris = datasets.load_iris()
X = iris.data
X_reduced = PCA(n_components=3).fit_transform(iris.data)
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
def plot_3d_distribution(X, y, title, fignum=1):
    fig = plt.figure(fignum, figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=110)
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y,
               cmap=plt.cm.Set1, edgecolor='k', s=40)
    ax.set_title(title)
    ax.set_xlabel("1st eigenvector")
    ax.w_xaxis.set_ticklabels([])
    ax.set_ylabel("2nd eigenvector")
    ax.w_yaxis.set_ticklabels([])
    ax.set_zlabel("3rd eigenvector")
    ax.w_zaxis.set_ticklabels([])
```

```python
estimators = [('k_means_3', KMeans(n_clusters=3)),
              ('k_means_3_bad_init', KMeans(n_clusters=3, n_init=1, init='random')),
              ('k_means_8', KMeans(n_clusters=8)),]
```

```python
fignum = 1
plot_3d_distribution(X_reduced, y, title='First three PCA directions', fignum=fignum)


for name, est in estimators:
    fignum += 1
    labels = est.fit_predict(X_reduced, y)
    plot_3d_distribution(X_reduced, labels, title=name, fignum=fignum)

```
