- 觀察將極端值資料直接刪除, 對於分布與迴歸分數的影響 (In[6], Out[6])

```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
%matplotlib inline

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_1stFlrSF(df, train_Y):
    plt.figure(figsize=(16, 5))
    sns.regplot(x = df['1stFlrSF'], y=train_Y, )
    plt.grid()
    plt.xticks(np.arange((min(df['1stFlrSF']) // 100 -1) * 100, max(df['1stFlrSF']) +1, 200))
    plt.show()
```

```python
# 顯示 1stFlrSF 與目標值的散佈圖
plot_1stFlrSF(df, train_Y)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

## 用上圖觀察結果，試將 1stFlrSF 限制在 400 和 2200之間 (調整離群值上下限)

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
df['1stFlrSF'] = df['1stFlrSF'].clip(400, 2200)

# 顯示 1stFlrSF 與目標值的散佈圖
plot_1stFlrSF(df, train_Y)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
keep_indexs = (df['1stFlrSF']> 400) & (df['1stFlrSF']< 2200)
df = df[keep_indexs]
train_Y = train_Y[keep_indexs]

# 顯示 1stFlrSF 與目標值的散佈圖
plot_1stFlrSF(df, train_Y)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

## 作業2-結論
使用**捨棄離群值**的方法，分數會在更好一點

可以觀察到，使用 **調整離群值** 方法時，回歸式會受上限值的資料點位影響，但那些經過上下限調整後的資料，並不屬於真實資料，造成擬和與實際資料不完全相符
